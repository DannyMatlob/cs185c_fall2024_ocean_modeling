When running the interpolation for the external forcing conditions, some grids had the incorrect number of values, so we must trim them down and save them.




We will import libraries and trim down to an array size of 84096000, but this number will change depending on the grid you specify if these notebooks change

```python
import os
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import netCDF4 as nc4
import cmocean.cm as cm
```

```python
input_dir = "D:\\project\\input"
output_dir = "."

array_size = 84096000
years = [2015,2016,2017]
```

```python
# make a file dictionary to loop over
file_prefix_dict = {'ATEMP':'EIG_tmp2m_degC_plus_ECCO_v4r1_ctrl',
             'AQH':'EIG_spfh2m_plus_ECCO_v4r1_ctrl',
             'SWDOWN':'EIG_dsw_plus_ECCO_v4r1_ctrl',
             'LWDOWN':'EIG_dlw_plus_ECCO_v4r1_ctrl',
             'UWIND':'EIG_u10m',
             'VWIND':'EIG_v10m',
             'PRECIP':'EIG_rain_plus_ECCO_v4r1_ctrl'}

variable_names = list(file_prefix_dict.keys())

variable_names
```

```python
for year in years:
    for i in range(len(variable_names)):
        #try:
        variable_name = variable_names[i]
        print(variable_name, ", ", year)
        
        CA_exf_grid = np.fromfile(os.path.join(input_dir,'exf',variable_name+'_'+str(year)),'>f4')
        CA_exf_grid = CA_exf_grid[:array_size]
    
        output_file = os.path.join(output_dir,'exf',variable_name+'_'+str(year))
        CA_exf_grid.ravel('C').astype('>f4').tofile(output_file)
```
